```python
import json
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

image_path = '/root/NeuralBabyTalk/data/noc_coco/val2014/COCO_val2014_000000083658.jpg'
json_path = '/root/NeuralBabyTalk/eval_results/noc_1_test.json'

with open(json_path, 'r') as json_file:
    data = json.load(json_file)

image_id = 83658 
caption = ''
for item in data:
    if item['image_id'] == image_id:
        caption = item['caption']
        break

img = mpimg.imread(image_path)
plt.imshow(img)

plt.title(caption)
plt.axis('off')
plt.show()
```
